```python
def violin(col,log=False):
    print('Showing info for: ', col)
    print('Number of missing values: ',train[col].isnull().sum())
    print('Number of zero values: ',sum(train[col]==0),)
    if log==True:
        sns.violinplot(x=train.Responders.values, y=np.log(1+train[col]).values)
    else:
        sns.violinplot(x="Responders", y=col, data=train, split=True)
```
